```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("C:/Users/Aashi Khanna/Downloads/Layoffs.fyi  - Tech Layoffs Tracker.csv")
df
```

```python
df.drop(columns=['Source'], inplace=True)
```

```python
df
```

```python
print(df.dtypes)
```

```python
# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Keep only records from 2025
df = df[df['Date'].dt.year == 2025]

# Create 'Month' column in format: '2025-01', '2025-02', etc.
df['Month'] = df['Date'].dt.strftime('%B') 

df
```

```python
# Remove '%' symbol and convert to float
df['%'] = df['%'].str.replace('%', '', regex=False).astype(float)
df['%'] = df['%'].fillna(0.0)

# Remove '$' symbol from '$ Raised (mm)' and convert to float
df['$ Raised (mm)'] = (
    df['$ Raised (mm)']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '')  # Optional: remove commas if present
    .astype(float)
)
df['$ Raised (mm)'] = pd.to_numeric(df['$ Raised (mm)'], errors='coerce').fillna(0.0)

df['# Laid Off'] = df['# Laid Off'].fillna(0)
df['Industry'] = df['Industry'].fillna("Unknown")
df['Stage'] = df['Stage'].fillna("Unknown")
df['Location HQ'] = df['Location HQ'].fillna("Unknown")
df['Country'] = df['Country'].fillna("Unknown")

df
```

```python
df['Layoff Region'] = df['Country'].apply(lambda x: 'US' if str(x).strip().upper() == 'UNITED STATES' else 'Non-US')
df
```

```python
df.to_csv("2025_Tech_Layoffs_Cleaned.csv", index=False)
```
